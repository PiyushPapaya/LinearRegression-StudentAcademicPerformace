#### Import libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

#### Import Database

```python
filepath = "/kaggle/input/student-academic-performance-trends/student_exam_scores.csv"
data = pd.read_csv(filepath)
data.head()
```

#### Define Label and Features

```python
features = ["hours_studied", "sleep_hours", "attendance_percent", "previous_scores"]
X = data[features]
y = data["exam_score"]
```

#### Split the Dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 1)
```

#### Define the ML - Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

#### Train the Model

```python
lr.fit(X_train,y_train)
```

#### Let the Model make Predictions

```python
y_pred = lr.predict(X_test)
```

#### Calculate the MAE and R2 Score

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"MAE: {mae}, R2 Score: {r2}")
```

#### Print the Weight and Bias

```python
feat = lr.coef_
bias = lr.intercept_
print("Weights of features:")
for i in range (4):
    print(f"{features[i]} : {feat[i]}")
print(f"Bias: {bias}")
```
